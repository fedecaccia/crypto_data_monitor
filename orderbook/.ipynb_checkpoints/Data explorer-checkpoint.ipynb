# Data explorer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data0 = pd.read_csv("orderbookOkexETHUSDT.csv", parse_dates=True, index_col="datetime")
data1 = pd.read_csv("orderbookKrakenETHUSD.csv", parse_dates=True, index_col="datetime")
f0 = 0.0015
f1 = 0.0026
```

```python
data0.head()
```

```python
plt.figure(1, figsize=(20,10))
plt.plot(data0["bid_val_0"])
plt.plot(data1["ask_val_0"])
plt.show()
```

```python
plt.figure(1, figsize=(20,10))
plt.plot(data0["ask_val_0"])
plt.plot(data1["bid_val_0"], color="red")
plt.show()
```

```python
# resample
max_delay = "60s"
pair0_ask = data0["ask_val_0"].resample(max_delay).mean().dropna()
pair1_ask = data1["ask_val_0"].resample(max_delay).mean().dropna()
pair0_bid = data0["bid_val_0"].resample(max_delay).mean().dropna()
pair1_bid = data1["bid_val_0"].resample(max_delay).mean().dropna()

# unique index between pairs

pair0_ask = pair0_ask[pair0_ask.index.isin(pair1_ask.index)]
pair1_ask = pair1_ask[pair1_ask.index.isin(pair0_ask.index)]
pair0_bid = pair0_bid[pair0_bid.index.isin(pair1_bid.index)]
pair1_bid = pair1_bid[pair1_bid.index.isin(pair0_bid.index)]
```

```python
p1 = -pair0_ask*(1+f0)+pair1_bid*(1-f1)
```
